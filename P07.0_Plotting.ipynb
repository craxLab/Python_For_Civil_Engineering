### _The anatomy of matplotlib plot !_

**_Figure_** : represents the overall window or page that everything is drawn on.

**_Axes_** : the area associated with ticks, labels etc., that data is plotted with functions. 

![](./Figures/figure-exp.png)

```python
x = [1,2,3,4,5,6,7,8,9]
y = [3,5,7,2,4,8,1,6,9]

plt.plot(x,y, color="k", linewidth=10)
plt.xlabel("Zaman")
plt.ylabel("Hız")
plt.title("Hız-zaman grafiği")
plt.xlim(0,10)
plt.ylim(0,15)
```

```python
import numpy as np

x = [1,2,3,4,5,6,7,8,9]
y1 = [3,5,7,2,4,8,1,6,9]
y2 = [1,5,6,3,2,7,2,6,7]

plt.figure(figsize=(5,3))
plt.plot(x,y1, color="k", linewidth=3)
plt.plot(x,y2, color="r", linewidth=3)
plt.legend(["Ferrari", "Volvo"], loc="upper right", fontsize=12)
plt.xlabel("Zaman")
plt.ylabel("Hız")
plt.title("Hız-zaman grafiği")
plt.xlim(0,10)
plt.xticks(np.arange(0,11,1))
plt.ylim(0,15)
plt.yticks(np.arange(0,16,1))
plt.grid()
plt.show()
```

## **Subplot** in matplotlib

A "figure" represents the whole window in the user interface, meanwhile, there can be **"subplots"** in different sizes. These **subplots** can be arranged in regular grid with the specified number of colomns and rows. 

![](./Figures/subplot.png)


```python
dataX = np.arange(20)
dataY1 = np.random.random(20)
dataY2 = np.random.random(20)

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(5,5))

ax[0].plot(dataX, dataY1, color="b", linewidth=2, linestyle=":", marker="*")
ax[0].set_title(" X grafiği Başlığı", fontsize=20)
ax[0].legend(["X grafiği"])
ax[0].set_xlabel("X değerleri")
ax[0].set_ylabel("Y1 değerleri")
ax[0].tick_params(axis="x")
ax[0].tick_params(axis="y")
ax[0].set_xlim(0,20)
ax[0].set_xticks(np.arange(0,21,2))
ax[0].set_ylim(0,1)
ax[0].set_yticks(np.arange(0,1.1,0.2))
ax[0].grid()

ax[1].plot(dataX, dataY2, color="r", linewidth=3, linestyle="-.", marker="v")
ax[1].set_xlabel("X değerleri", fontsize=20)
ax[1].set_ylabel("Y2 değerleri", fontsize=20)
ax[1].tick_params(axis="x")
ax[1].tick_params(axis="y")
ax[1].set_xlim(0,20)
ax[1].set_xticks(np.arange(0,21,2))
ax[1].set_ylim(0,1)
ax[1].set_yticks(np.arange(0,1.1,0.2))
ax[1].grid()

fig.tight_layout()
plt.show()
```
