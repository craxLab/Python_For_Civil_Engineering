## Elastic Horizontal Acceleration Design Spectrum according to TEC-2018

***Create period (T) axis:***

```python
T = range(1,21)
print(T)
```

```python
T = range(1,21,0.1)
print(T)
```

```python
import numpy as np
T = np.arange(0.01, 20, 0.1)
print(T)
```

**Calculation of Acceleration Design Spectrum:**

```python
Ss = 1.518 ; Fs = 0.800 ; S1 = 0.417 ; F1 = 2.366

Sds = Ss*Fs
Sd1 = S1*F1
Ta = 0.2*Sd1/Sds
Tb = Sd1/Sds
Tl = 6

Sae = []
for Ti in T:
    if Ti > 0 and Ti<=Ta:
        Saei = (0.4+0.6*Ti/Ta)*Sds
    elif Ti > Ta and Ti<=Tb:
        Saei = Sds
    elif Ti > Tb and Ti<=Tl:
        Saei = Sd1/Ti
    else:
        Saei = Sd1*Tl/Ti**2
    Sae.append(Saei)

print(Sae)
```

```python
import matplotlib.pyplot as plt
plt.plot(T,Sae)
```

# Example 2B:

## Elastic Horizontal Displacement Design Spectrum according to TEC-2018

![](./Figures/DesignSpectrum/Ex_Design_SpectrumHor3.png)

```python
g = 9.81
pi = 3.14

Sde = []
for i in range(len(T)):
    Sdei = T[i]**2/(4*pi**2)*g*Sae[i]
    Sde.append(Sdei)

print(Sde)
```

```python
import matplotlib.pyplot as plt
plt.plot(T,Sde)
```

Note: The length of the lists should be equal.
