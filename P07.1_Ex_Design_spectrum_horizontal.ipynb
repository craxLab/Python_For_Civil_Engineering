## Example: Horizontal Acceleration Design Spectrum according to TSCB-2019

```python
# TBDY 2018'e göre "Yatay Elastik Tasarım Spektrumu" hesabı
import numpy as np

# INPUTS
Ss = 1.518 ; Fs = 0.800 ; S1 = 0.417 ; F1 = 2.366

# CALCULATIONS
Sds = Ss*Fs
Sd1 = S1*F1
Ta = 0.2*Sd1/Sds 
Tb = Sd1/Sds
Tl = 6.0

# define a function for spectrum value calculation
def spec(Ti):
    if Ti <= Ta:
        Saei = (0.4 + 0.6*Ti/Ta)*Sds
    elif Ti > Ta and Ti <= Tb:
        Saei = Sds
    elif Ti > Tb and Ti <= Tl:
        Saei = Sd1/Ti
    else: 
        Saei = Sd1*Tl/Ti**2
        
    return Saei


# Calculate all Sae values for any  given T range
T = np.arange(0.01, 7.1, 0.1)

Sae = list()    # [g]; elastik tasarım spektral ivmeleri
for Ti in T:
    Saei = spec(Ti)
    Sae.append(Saei)

print("T=", T) 
print("Sae=", Sae)
```

**Plot Spectrum**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))
plt.plot(T,Sae, color="k", linewidth=3)
plt.xlabel("Period (s)")
plt.ylabel("Sae (g)")
plt.title("Yatay elastik tasarım spektrumu")
plt.xlim(0,7)
plt.xticks(np.arange(0,8,1))
plt.ylim(0,1.5)
plt.yticks(np.arange(0,1.6,0.25))
plt.grid()
plt.show()
```

**Plot Spectra**

```python
import matplotlib.pyplot as plt
Sae = np.array(Sae)

plt.figure(figsize=(5,3))
plt.plot(T,Sae, color="k", linewidth=3)
plt.plot(T,Sae*9.81, color="r", linewidth=3)
plt.legend(["g","m/s2"])
plt.xlabel("Period (s)")
plt.ylabel("Sae (g)")
plt.title("Yatay elastik tasarım spektrumu")
plt.xlim(0,7)
plt.xticks(np.arange(0,8,1))
plt.ylim(0,15)
plt.yticks(np.arange(0,16,2.5))
plt.grid()
plt.show()
```
